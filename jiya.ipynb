```python
import pandas as pd
```

```python
df = pd.read_csv("data/최종_최종__월별기상관측.csv",encoding="cp949")
df
```

```python
df.drop(df.columns[:1],axis=1,inplace=True)   # 0,1번째 컬럼 drop
df.head()
```

```python
gangwon_spring=pd.read_csv("data/gangwon_spring.csv")
gangwon_spring
```

```python
#강원도 봄재배, 봄 기준: 3월~5월
```

```python
gangwon1=df[df['행정구역']=='강원도']
gangwon1.head()
```

```python
gangwon2 = pd.pivot_table(gangwon1, index=["일시", '행정구역'],aggfunc="median")
gangwon2
```

```python
gangwon_spring_wt=gangwon2.query('일시.str.contains("03|04|05")', engine='python')
gangwon_spring_wt
```

```python
gangwon_spring_wt_2000=gangwon_spring_wt.query('일시.str.contains("2000")', engine='python')
gangwon_spring_wt_2000
```

```python
gangwon_spring_wt_2000_total = pd.pivot_table(gangwon_spring_wt_2000, index=[ '행정구역'],aggfunc="median")
gangwon_spring_wt_2000_total.reset_index(inplace=True)#데이터 프레임 전환
gangwon_spring_wt_2000_total.insert(0,'연도','2020')
gangwon_spring_wt_2000_total.insert(1,'계절','봄')
```
