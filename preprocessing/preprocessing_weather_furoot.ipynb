```python
def fill_tem(temperature, dataframe, pivot) : 
    for i in range(len(dataframe)) : 
        dt = dataframe.loc[i,"일시"]
        dstr = dataframe.loc[i,"행정구역"]

        if np.isnan(dataframe.loc[i,temperature]) :  # 기온 관련 컬럼값이 NaN일 때 
            dataframe.loc[i,temperature] = pivot[dstr][dt]   # 동일 일시, 동일 행정 구역의 중앙값을 채움
    
    return dataframe
```

```python
yr_list= np.arange(2000, 2022)
detail = ["-01","-02","-03","-04","-05","-06","-07","-08","-09","-10","-11","-12"]
spring= []
summer= []
fall= []
winter= []

for yr in yr_list:
    # 3월부터 5월까지
    for i in range(2,5):
        val= str(yr) + detail[i]
        spring.append(val)
    # 5월부터 9월까지
    for i in range(4,9):
        val= str(yr) + detail[i]
        summer.append(val)
    # 8월부터 11월까지
    for i in range(7,11):
        val= str(yr) + detail[i]
        fall.append(val)
    # 8월부터 다음 1월까지
    # range(11,12) + range(0,3)
    for i in range(10,12):
        val= str(yr) + detail[i]
        winter.append(val)
    for i in range(0,1):
        val= str(yr+1) + detail[i]
        winter.append(val)


# season 배열 생성
season = ["spring", "summer", "fall", "winter"]
season_list= [spring, summer, fall, winter]

winter

```
